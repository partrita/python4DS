# 사전 필요 조건

데이터 과학 여정을 시작하신 것을 축하합니다! 이 장에서는 데이터 과학을 배우고 수행하는 데 필요한 도구를 설치하거나 액세스하는 데 도움을 드릴 것입니다. 이 책을 최대한 활용하기 위해 이미 알고 계실 것으로 가정하는 몇 가지 사항이 있습니다. 일반적으로 숫자에 대한 이해가 있어야 하며 프로그래밍 경험이 이미 있다면 도움이 됩니다.

## 소개

다음과 같은 설정에 도움을 드릴 것입니다.

- *통합 개발 환경*(IDE), 코드를 작성하고 실행하는 곳
- *파이썬* 설치, 컴퓨터가 파이썬 코드를 해석하고 실행할 수 있도록
- 파이썬의 기능을 확장하는 *패키지* 설치

방법에 대한 자세한 정보가 많이 있겠지만, 먼저 이것이 *무엇*인지 잠시 설명하겠습니다.

### 통합 개발 환경 또는 IDE

*통합 개발 환경*(IDE)은 코딩을 더 쉽게 만드는 몇 가지 도구를 제공하는 소프트웨어 응용 프로그램입니다. 이 중 가장 중요한 것은 코드 자체를 작성하는 방법입니다! IDE가 프로그래밍하는 유일한 방법은 아니지만 아마도 가장 유용할 것입니다. 언어와 언어를 작성하는 장소(IDE)는 별개라는 점에 유의하십시오. 언어는 지침을 처리하는 방법이고 IDE는 해당 지침을 작성하는 곳입니다. 시중에는 많은 통합 개발 환경(IDE)이 있습니다. 이 책은 모든 주요 운영 체제에서 작동하고 가장 인기 있는 것 중 하나인 Microsoft의 *Visual Studio Code*를 강력히 권장합니다. 다음은 *Visual Studio Code*가 제공하는 몇 가지 유용한 기능입니다.

- 코드를 대화형으로(줄 단위로) 또는 한 번에 실행하는 방법

- 코드에서 디버그(오류 찾기)하는 방법

- 일반적으로 사용되는 소프트웨어 패키지에 대한 유용한 정보에 빠르게 액세스하는 방법

- 코드가 모범 사례 지침을 따르도록 자동 코드 서식 지정

- <kbd>TAB</kbd>을 누를 때 코드 자동 완성

- 기본 오류에 대한 자동 코드 검사

- 코드 실행의 논리적 순서를 추적할 수 있도록 괄호를 쌍으로 색칠!

### 파이썬 인터프리터

파이썬은 읽을 수 있는 프로그래밍 언어이자 컴퓨터가 읽고 해석한 다음 이를 기반으로 지침을 수행할 수 있는 언어입니다. 컴퓨터가 파이썬 코드를 읽고 실행하려면 컴퓨터에 파이썬을 설치해야 합니다. 컴퓨터에 파이썬 "인터프리터"를 설치하는 방법은 여러 가지가 있지만 이 책에서는 유연성과 단순성 때문에 파이썬의 *uv 배포판*을 권장합니다.

### 패키지

파이썬 **패키지**는 설치된 파이썬 버전의 기능을 확장하는 함수, 데이터 및 설명서 모음입니다. 대부분의 필요한 기능이 추가 패키지에서 제공되므로 패키지를 사용하는 것은 대부분의 데이터 과학에 핵심적입니다. 많은 파이썬 코드 스크립트 시작 부분에 `import numpy as np`와 같은 문을 볼 수 있습니다. 이는 설치된 패키지(여기서는 `numpy`라는 패키지)를 사용하고 스크립트의 나머지 부분에서 편의를 위해 축약된 이름(`np`)을 지정하라는 지침입니다. 그런 다음 `numpy` 패키지의 함수는 `np.`와 같은 구문을 통해 액세스합니다. 예를 들어 `x`가 숫자를 포함하는 변수인 경우 `np.log(x)`를 사용하여 로그를 취할 수 있습니다. 패키지는 한 번만 설치하면 됩니다.

### 일반적인 워크플로

코드를 사용한 분석의 일반적인 워크플로는 다음과 같을 수 있습니다.

- *통합 개발 환경*(IDE)을 엽니다.
- *IDE*에서 스크립트(코드가 포함된 텍스트 파일)에 일부 코드를 작성합니다.
- 수행 중인 분석에 필요한 경우 추가 *패키지*를 설치합니다.
- *IDE*를 사용하여 스크립트의 코드 일부 또는 전체 스크립트를 *파이썬* 및 추가 기능 *패키지*에서 실행하고 결과를 표시합니다.

이 워크플로를 달성하는 두 가지 방법을 살펴보겠습니다.

1. 자신의 컴퓨터에 *IDE*, *파이썬* 및 추가 *패키지* 설치
2. 인터넷 브라우저를 통해 액세스하는 클라우드의 컴퓨터 사용. 클라우드 컴퓨터에는 *IDE*와 *파이썬*이 내장되어 있으며 추가 *패키지*도 쉽게 설치할 수 있습니다. 그러나 무료 버전은 월 60시간으로 제한됩니다.

더 편한 것을 선택하십시오! 결국에는 두 가지 모두 시도하게 될 것입니다.

## 자신의 컴퓨터에서 시작하는 방법

이 지침은 자신의 컴퓨터에서 로컬로 파이썬으로 작업하려는 경우를 위한 것입니다.

### 파이썬 설치

파이썬을 다운로드하고 설치하려면 모든 주요 운영 체제에서 사용할 수 있는 파이썬의 uv "배포판"을 사용합니다. 설치하려면 [이 웹사이트](https://docs.astral.sh/uv/getting-started/installation/#installation-methods)의 지침을 따르십시오. 일반 프로그램을 설치하는 것과 달리 *명령줄*을 사용하여 파이썬을 설치합니다. Linux, Mac 및 Windows에는 모두 내장 명령줄이 있습니다. Mac 또는 Linux에서는 *터미널*을, Windows에서는 *Powershell*을 검색합니다. 이러한 앱은 명령을 입력할 수 있는 상자를 표시합니다. 작성 당시 명령은 다음과 같습니다.

```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

Linux 및 Mac의 경우,

```powershell
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

Windows의 경우. Return 키를 눌러 명령을 실행합니다.

uv를 설치한 후에는 `uv --version`을 실행하여 제대로 설치되었는지 확인할 수 있습니다. "uv"와 최신 버전 번호가 표시된 메시지가 나타납니다.

아직 파이썬을 다운로드하지 않았습니다. 파이썬 환경을 관리할 **uv**를 다운로드했습니다. 따라서 다음을 실행해야 합니다.

```bash
uv python install
```

### 통합 개발 환경인 Visual Studio Code 설치

[Visual Studio Code](https://code.visualstudio.com/)는 모든 주요 운영 체제에서 사용할 수 있는 Microsoft의 무료 오픈 소스 IDE입니다. 파이썬 자체와 마찬가지로 Visual Studio는 패키지로 확장할 수 있으며, 이 경우 확장이라고 하는 이러한 패키지가 매우 유용합니다. 파이썬 외에도 Visual Studio Code는 수많은 다른 언어를 지원합니다.

Visual Studio Code를 다운로드하고 설치합니다. 도움이 필요하면 Visual Studio Code를 다운로드하고 설치한 다음 스크립트와 노트북 모두에서 파이썬 코드를 실행하는 방법을 안내하는 아래 비디오가 있습니다. 이 장의 나머지 부분에서 이러한 지침을 자세히 살펴보겠습니다.

노트북과 대화형 창을 테스트하는 경우 새 폴더를 만들고 Visual Studio Code에서 해당 폴더에서 작업해야 합니다. VS Code에서 폴더를 명시적으로 연 후 Visual Studio Code 내에서 명령줄을 열고 `uv init`를 실행하여 파이썬 설치를 만들고 `uv add jupyter`를 실행하여 이 초기 파이썬 환경에 대화형 창을 만들 수 있는 수단을 제공합니다.

## 클라우드에서의 데이터 과학

이 지침은 클라우드의 컴퓨터를 사용하여 원격으로 데이터 과학을 수행하려는 경우를 위한 것입니다.

클라우드에서 데이터 과학을 수행하는 방법은 여러 가지가 있지만 가장 간단한 방법을 알려 드리겠습니다. 이를 위해서는 [Github 계정](https://github.com/)에 가입해야 합니다. Github은 Microsoft가 소유한 조직으로 클라우드에 코드를 백업하는 방법과 클라우드 컴퓨팅을 포함한 다양한 서비스를 제공합니다. 제공되는 서비스 중 하나는 *Github Codespaces*입니다. GitHub Codespace는 브라우저 창에서 연결하는 온라인 클라우드 컴퓨터입니다. 월 60시간의 무료 컴퓨팅 시간을 제공합니다.

```{note}
Github Codespaces에서 무료 등급 시간을 초과하면 사용한 추가 GitHub Codespaces 시간에 대해 신용 카드로 요금이 청구됩니다.
```

Github 계정에 가입한 후 [Github Codespaces](https://github.com/codespaces)로 이동하여 "무료로 시작하기"를 클릭합니다. "빠른 시작 템플릿" 메뉴가 표시됩니다. "Jupyter Notebook"이라고 표시된 곳 아래에서 "이 템플릿 사용"을 누릅니다.

여러 패널이 있는 새 페이지가 로드됩니다. 이것은 자신의 컴퓨터에 설치한 것처럼 작동하는 Visual Studio Code의 온라인 버전입니다. 이미 파이썬 버전이 설치되어 있습니다. 터미널에서 `python --version`을 실행하여 어떤 버전인지 확인할 수 있습니다. 터미널은 일반적으로 Visual Studio Code의 가장 낮은 패널에 있으며 Codespaces에서는 일반적으로 환영 메시지가 표시됩니다.

그러나 Codespaces의 파이썬 배포판은 이 책에서 사용하는 **uv** 배포판과 다를 수 있습니다. 하지만 codespaces 컴퓨터에 **uv**를 설치하고 위 "파이썬 설치" 섹션의 다른 지침을 따를 수 있습니다.

## 첫 파이썬 코드 실행하기

### Visual Studio Code 익숙해지기

Visual Studio Code를 설치하고 열면(자신의 컴퓨터 또는 클라우드에서) 화면 왼쪽 세로 아이콘 막대에서 '확장 프로그램' 탭(사각형 4개처럼 보이는 아이콘)으로 이동합니다. VS Code의 확장 프로그램 패널 내 텍스트 상자를 사용하여 검색할 수 있는 *VS Code용 파이썬 확장 프로그램*을 설치해야 합니다. 클라우드 버전을 사용하는 경우 이미 설치되어 있을 수 있습니다.

설치해야 할 유용한 다른 확장 프로그램(이미 설치되어 있지 않은 경우)이 몇 가지 있습니다.

- Jupyter
- Pylance
- indent-rainbow

아직 가지고 놀 파이썬 코드나 해당 파이썬 코드를 실행할 대화형 창이 없지만 Visual Studio Code의 *일반적인* 보기의 다른 부분에 대해 잠시 익숙해지는 것이 좋습니다.

![Visual Studio Code의 일반적인 사용자 보기](https://github.com/aeturrell/coding-for-economists/blob/main/img/vscode_layout.png?raw=true)

위 그림은 파이썬 세션이 실행 중이고 파이썬 스크립트가 열려 있는 Visual Studio Code의 일반적인 레이아웃을 보여줍니다. 맨 왼쪽에 있는 긴 세로 패널은 패널 1과 2에 표시되는 내용을 변경합니다. 현재 파일 탐색기가 선택되어 있습니다. 그림의 번호가 매겨진 부분을 살펴보겠습니다.

1. 1과 2 왼쪽에 있는 아이콘에서 탐색기 옵션을 선택하면 현재 열려 있는 폴더의 내용이 1에 표시됩니다.
2. 이것은 3에서 열려 있는 파일의 주요 부분에 대한 개요입니다.
3. 이것은 멋진 텍스트 편집기일 뿐입니다. 위 그림에서는 파이썬 스크립트(코드가 포함되어 있고 이름이 `.py`로 끝나는 파일)를 보여줍니다. 곧 코드를 선택하고 <kbd>Shift</kbd> + <kbd>Enter</kbd>('Enter'는 일부 키보드에서는 'Return'으로 표시됨)를 누르면 패널 5에 결과가 표시되는 코드가 실행되는 것을 볼 수 있습니다.
4. 이것은 컴퓨터가 실행할 명령을 입력할 수 있는 명령줄 또는 *터미널*입니다. 명령을 시도하려면 `date`(Mac/Linux) 또는 `date /t`(Windows)를 입력합니다. 여기서 추가 *패키지*를 설치합니다.
5. 이것은 대화형 파이썬 창으로, 스크립트에서 선택하고 실행한 후(3 참조) 코드와 코드 출력이 표시되는 곳입니다. 실행한 코드와 해당 실행의 모든 출력을 보여줍니다. 표시된 스크린샷에서는 코드가 플롯을 만들었습니다. 사용 중인 파이썬의 이름과 버전은 대화형 창 상단에 나타납니다.

창 하단의 파란색 막대에는 현재 VS Code에서 사용 중인 파이썬 버전을 포함하여 많은 유용한 정보가 배열되어 있습니다.

### 파이썬 코드 실행하기

이제 첫 코드를 만들고 실행합니다. 막히면 [VS Code 설명서](https://code.visualstudio.com/docs/python/python-tutorial)에 더 자세한 튜토리얼이 있습니다.

Visual Studio Code에서 화면 왼쪽에 있는 "탐색기" 기호(일부 파일)를 클릭하여 파일 탐색기를 엽니다. 컴퓨터에서 작업을 시도하기에 좋은 위치에 있는지 확인하고 그렇지 않은 경우 만족할 때까지 파일 -> 폴더 열기를 사용하여 현재 폴더를 변경합니다.

이제 Visual Studio Code *내에서* 터미널을 엽니다. uv를 설치한 터미널과 동일한 유형인지 확인하십시오. 예를 들어 Windows에서는 Powershell을 사용합니다(다른 유형의 터미널을 사용하도록 선택할 수 있음). 터미널 패널을 여는 바로 가기는 Linux 및 Windows에서는 Ctrl + \`, Mac에서는 Cmd + \`입니다. 터미널 패널은 VS Code 하단에 나타납니다. 터미널에서 `uv init`를 실행하여 파이썬 설치를 시작합니다. **이 파이썬 설치는 현재 폴더에만 존재합니다**. 그런 다음 이 섹션에 필요한 패키지를 설치하려면 `uv add jupyter`를 실행합니다. (터미널과 패키지 설치에 대해서는 곧 훨씬 더 자세히 다룰 것입니다.)

이제 왼쪽 탐색기 패널이 열려 있는 상태에서 "+" 기호가 있는 빈 종이처럼 보이는 기호를 클릭합니다. 이렇게 하면 새 파일이 만들어지고 커서가 파일 이름을 지정하도록 이동합니다. 이름을 `hello_world.py`로 지정합니다. 파일 확장자 `.py`는 Visual Studio Code에 이것이 파이썬 스크립트임을 암시적으로 알려주므로 매우 중요합니다.

Visual Studio Code 편집기에서 파일에 한 줄을 추가합니다.

```python
print('Hello World!')
```

자, 이제 실제로 **코드를 실행**해 볼까요? 파일에 입력한 `print("Hello world!")` 텍스트를 선택/강조 표시하고 마우스 오른쪽 버튼을 클릭합니다. 여기에 많은 옵션이 표시되지만 원하는 옵션은 **"대화형 창에서 선택/줄 실행"**입니다.

이렇게 하면 Visual Studio Code 내에 새 '대화형' 패널이 나타나고 짜잔, 다음과 같이 표시됩니다.

```python
print("Hello world!")
```

*대화형 창*은 스크립트에서 열려 있거나 대화형 창 코드 상자에 직접 입력한 코드를 실행하는 편리하고 유연한 방법입니다. 대화형 창은 스크립트의 일부 줄을 실행했든 직접 입력했든 할당된 모든 변수(예: `x = 5`와 같은 코드 문)를 '기억'합니다. 대화형 창으로 작업하는 것은 Stata, Matlab 또는 R을 사용해 본 사람이라면 누구나 익숙할 것입니다. 미리 전체 스크립트를 처음부터 끝까지 작성할 필요가 없습니다. 대신 코드를 변경하고 줄 단위로 (다시) 실행하면서 즉흥적으로 작업할 수 있습니다.

코드를 실행할 때마다 마우스 오른쪽 버튼을 클릭해야 한다면 번거로우므로 강조 표시된 코드를 대화형 창으로 보내 실행하는 *키보드 바로 가기*를 만들 것입니다. 이렇게 하려면 다음을 수행합니다.

- Visual Studio Code 구성 메뉴(왼쪽 하단의 톱니바퀴)를 엽니다.
- 설정으로 이동합니다.
- 상자에 "jupyter send"를 입력하여 "대화형 창 > 텍스트 편집기: 선택 실행" 항목이 나타나도록 합니다.
- 이 항목 옆의 확인란이 선택되어 있는지 확인합니다.

이제 스크립트로 돌아가서 `print("Hello world!")`가 있는 줄에 커서를 놓고 <kbd>Shift</kbd>+<kbd>Enter</kbd>를 누릅니다. "Hello world!"가 다시 나타나지만 이번에는 훨씬 쉬워졌습니다.

```{admonition} 대신 터미널에서 코드 실행하기
:class: dropdown

대화형 창만이 코드를 실행하는 유일한 방법은 아닙니다. 터미널에서도 수행할 수 있습니다. 데이터 과학에는 덜 인기가 있지만 때때로 유용합니다. 이렇게 하려면 선택한 코드에서 마우스 오른쪽 버튼을 클릭하고 "파이썬 실행 -> 터미널에서 선택/줄 실행"을 선택합니다.
```

*대화형 창*을 더 많이 활용해 보겠습니다. 하단에는 '여기에 코드를 입력하고 Shift-Enter를 눌러 실행'이라는 상자가 있습니다. 스크립트에서 줄을 실행하는 것과 동일한 효과를 얻으려면 여기에 `print('Hello World!')`를 직접 입력합니다. 또한 대화형 창에서 실행하는 모든 변수(스크립트에서 또는 상자에 직접 입력하여)는 유지됩니다.

변수가 어떻게 유지되는지 보려면 대화형 창의 코드 입력 상자에 `hello_string = 'Hello World!'`를 입력하고 Shift-Enter를 누릅니다. 이제 `hello_string`을 입력하고 Shift+Enter를 누르면 방금 만든 변수의 내용이 표시됩니다. 대화형 창 상단(중지 기호와 파일 저장 기호 사이)의 그리드 기호를 클릭할 수도 있습니다. 이것은 변수 탐색기이며 이 대화형 세션에서 만든 모든 변수를 보여주는 패널이 나타납니다. `str` 유형의 `hello_string`이라는 변수가 `Hello World!` 값으로 표시되어야 합니다.

이것은 대화형 창으로 작업하는 두 가지 방법, 즉 스크립트에서 (세그먼트) 실행 또는 입력 상자에 직접 코드 작성을 보여줍니다. 변수를 어떤 방식으로 입력했든 대화형 창의 해당 세션 내에서 모두 기억됩니다.

```{admonition} 프로젝트 디렉터리 내에서 대화형 창 및 터미널 시작하기
:class: dropdown
Visual Studio Code에서 설정 메뉴에서 "Jupyter: Notebook 파일 루트"를 `${workspaceFolder}`로 설정하여 대화형 창이 프로젝트의 루트 디렉터리에서 시작되도록 할 수 있습니다. 통합 명령줄의 경우 "터미널 > 통합: Cwd"도 `${workspaceFolder}`로 변경합니다.
```

그리고 엔터 키를 누릅니다. 위에서 `패키지이름`은 예를 들어 `pandas`일 수 있습니다. 설치에 문제가 있는 경우 인터넷에 연결되어 있는지, 방화벽이나 프록시에서 [PyPI](https://pypi.org/)(파이썬 패키지 인덱스)가 차단되지 않았는지 확인합니다. 이 방법으로 데이터 분석 패키지 **polars**를 설치해 볼 수 있습니다. `uv add polars`를 실행합니다. 이후 장에서 **polars**를 사용하는 방법을 살펴보겠지만 제대로 설치되었는지 확인하려면 "polars가 성공적으로 설치되었습니다"라는 메시지와 버전 번호가 표시되는지 확인하면 됩니다.

명령줄에 `uv pip list`를 입력하여 이전에 설치한 패키지를 볼 수 있습니다.

{ref}`workflow-packages-and-environments`에서 패키지 설치 및 사용에 대한 자세한 내용을 확인할 수 있습니다.

설치에 문제가 있는 경우 인터넷에 연결되어 있는지, 방화벽이나 프록시에서 [PyPI](https://pypi.org/)(파이썬 패키지 인덱스)가 차단되지 않았는지 확인합니다.

## 이 책의 코드를 실행하는 다른 방법

자신의 컴퓨터나 Github Codespaces를 통해 클라우드에서 이 책을 따르는 것 외에도 몇 가지 다른 옵션을 통해 온라인으로 코드를 실행할 수 있습니다. 첫 번째 방법이 가장 시작하기 쉽습니다.

1. [Google Colab 노트북](https://research.google.com/colaboratory/). 대부분 무료로 사용할 수 있습니다. 페이지 상단의 로켓 기호 아래에 있는 'Colab' 버튼을 사용하여 이 책의 대부분의 페이지를 대화형으로 실행할 수 있습니다. 스크립트(.py 파일)가 아닌 노트북(코드와 텍스트 혼합) 형식으로 제공되지만 작성하는 코드는 동일합니다. 패키지를 최신 버전으로 업데이트해야 할 수 있습니다. Colab에서는 코드 셀에서 `!pip install **패키지이름**`을 실행하여 이 작업을 수행할 수 있습니다. 추가 느낌표는 Colab에 이것이 파이썬이 아닌 운영 체제에 대한 지침임을 알려줍니다.
2. [Gitpod 작업 공간](https://www.gitpod.io/). Codespaces의 대안입니다. 이것은 파이썬이 설치된 Visual Studio Code의 원격 클라우드 기반 버전이며 파이썬 스크립트를 실행합니다. 무료 등급은 월 50시간을 지원합니다.
